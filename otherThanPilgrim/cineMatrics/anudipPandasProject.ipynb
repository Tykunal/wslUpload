```python
import pandas as pd
import matplotlib.pyplot as plt
import json
```

```python
df=pd.read_csv("movie_data.csv")
# df


# print(data)
df.shape
```

```python
df["release_date"]=pd.to_datetime(df["release_date"])
dff=df[df["release_date"]>'2015-01-01']
df=pd.DataFrame(dff)
df.shape
```

```python
df=df.drop(columns=["status","homepage","id"])
df.shape
```

```python
high_budget=df.sort_values(by="budget",ascending=False)
high_budget=high_budget.head(50)
high_budget
```

```python
plt.figure(figsize=(8, 10))
# budgetList = [280000000,75000000,150000000,108000000,70000000]

# countBudget = {}

# for i in df["budget"]:
#     if i in countBudget:
#         countBudget[i]+=1
#     else:
#         countBudget[i] = 1

# print(countBudget)

# plt.xticks(budgetList, rotation=90)
# plt.xticks(budgetList)
plt.barh(high_budget['original_title'],high_budget["budget"],color='red')
plt.gca().invert_yaxis()
```

```python
high_revenue=df.sort_values(by="revenue",ascending=False)
high_revenue=high_revenue.head(50)

plt.figure(figsize=(8, 10))
plt.barh(high_revenue['original_title'],high_revenue["revenue"],color='blue')
plt.gca().invert_yaxis()
```

```python
budgetVsRevenue=df.head(50)
plt.figure(figsize=(12, 10))

plt.scatter(budgetVsRevenue['budget'], budgetVsRevenue['revenue'], color='b', alpha=0.6)


plt.title('Budget vs Revenue')
plt.xlabel('Budget ($)')
plt.ylabel('Revenue ($)')

for i, row in budgetVsRevenue.iterrows():
    plt.annotate(row['title'], (row['budget'], row['revenue']), fontsize=8, alpha=0.7)

plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

```python
highPopularity = df.sort_values(by="popularity", ascending=False).head(50)

plt.figure(figsize=(10, 12))

plt.barh(highPopularity["title"], highPopularity["popularity"], color='green')

plt.xlabel("Popularity")
plt.ylabel("Movie Titles")
plt.title("Top 50 Most Popular Movies")

plt.gca().invert_yaxis()

plt.show()


```

```python
highRating = df.sort_values(by="vote_average", ascending=False).head(20)
plt.figure(figsize=(12, 8))
plt.barh(highRating['title'], highRating['vote_average'], color='purple')
plt.xlabel('Rating')
plt.ylabel('Movie Titles')
plt.title('Top 20 Movies by Rating')
plt.gca().invert_yaxis()
plt.show()
```

```python

ratingVsPopularity=df.head(20)
plt.figure(figsize=(10, 6))

plt.scatter(ratingVsPopularity['vote_average'], ratingVsPopularity['popularity'], color='b', alpha=0.6)

plt.title('Vote Average vs Popularity')
plt.xlabel('Votes Average')
plt.ylabel('Popularity')

for i, row in ratingVsPopularity.iterrows():
    plt.annotate(row['title'], (row['vote_average'], row['popularity']), fontsize=8, alpha=0.7)

plt.grid(True, linestyle='--', alpha=0.5)

```

```python
highPopularity = df.sort_values(by = "popularity", ascending = False).head(50)
data = list(highPopularity["genres"])
# print(highPopularity["title"])
# highPopularity
genrePopularity = {}


for d in data:
    genres_list = json.loads(d)
    genre_names = [genre['name'] for genre in genres_list]
    for i in range(len(genre_names)):
        if genre_names[i] in genrePopularity:
            genrePopularity[genre_names[i]] +=1
        else:
            genrePopularity[genre_names[i]] = 1    
    # for g in genre_names:
    #     genrePopularity[g] +=1
    # print(genre_names)
print(genrePopularity)

sorted_genrePopularity = dict(sorted(genrePopularity.items(), key=lambda item: item[1], reverse=True))

plt.figure(figsize=(12,8))
wedges, texts, autotext = plt.pie(sorted_genrePopularity.values(), autopct="%1.1f%%", startangle=90)
plt.legend(wedges, sorted_genrePopularity.keys(), title="Genres", loc="center left", bbox_to_anchor=(-0.2, 0.5), fontsize=10)

plt.axis('equal') 
plt.show()
```
